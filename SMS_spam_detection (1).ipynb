```python
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
df = pd.read_csv("spam.csv", encoding='ISO-8859-1')
df = df.drop(["Unnamed: 2", "Unnamed: 3", "Unnamed: 4"], axis=1)
df.columns = ["label", "message"]
```

```python
df.drop_duplicates(inplace =True)
df.duplicated().sum()
```

```python
df.groupby('label').describe()
```

```python
df["spam_coding"] = df["label"].apply(lambda x : 1 if x =="spam" else 0)
```

```python
df.head(3)
```

```python
x_train , x_test , y_train , y_test = train_test_split(df.message , df.spam_coding)
```

```python
count_words = CountVectorizer()
x_train_word_count = count_words.fit_transform(x_train)
```

```python
model = MultinomialNB()
model.fit(x_train_word_count, y_train)
```

```python
#testing for not spam
test_email = ["Hey how was your vacation?"]
ham_email_count = count_words.transform(test_email)
confidence = model.predict_proba(ham_email_count)
```

```python
if model.predict(ham_email_count) == 0:
        print(f"Prediction: not spam - Confidence: {confidence[0][1]:.2f} - Message: {test_email[0]}")
```

```python
#testing for spam
spam_email = ["you won IPHONE 15 PRO MAX click here to claim"]
spam_email_count = count_words.transform(spam_email)
confidence = model.predict_proba(spam_email_count)
```

```python
if model.predict(spam_email_count) == 1:
        print(f"Prediction: Spam - Confidence: {confidence[0][1]:.2f} - Message: {spam_email[0]}")
```

```python
x_test_word_count = count_words.transform(x_test)
```

```python
model.score(x_test_word_count, y_test)
```
